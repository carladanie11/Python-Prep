3) Para los valores enteros del 1 al 20, imprimir por pantalla si es par o impar

```python
#Resuelto con buble For
for i in range(1, 21):
    if i % 2 == 0:
        print('El numero', str(i), 'es par')
    else:
        print('El numero', str(i), 'es impar')

```

```python
# Resuelto con bucle while
i = 1
while(i <= 20):
    if(i % 2 == 0):
        print(f'El numero {i} es par')
    else:
        print(f'El numero {i} es impar')
    i = i + 1        # i+=1 es lo mismo
```

6) Utilizar un ciclo while para realizar el factorial de un número guardado en una variable, sólo si la variable contiene un número entero mayor a 0

```python
#factorial
# 5! = 5*4*3*2*1  cinco factorial
# 4! = 4*3*2*1
# 3! = 3*2*1
# 2! = 2*1
```

```python
# iteracion 1:
# n = 5
# n-1 = 4
# factorial = 5 * 4

#iteracion 2:
# n = 4
# n-1 = 3
# factorial = 5*4*3

#iteracion 3:
# n=3
# n-1 = 2
# factorial = 5*4*3*2

```

```python
#Resuelto con bucle while
n = 5
if(type(n) == int):
    if(n > 0):
        factorial = n
        while(n > 2):
            n = n - 1
            factorial = factorial * n
        print('El factorial es ', factorial)
    else:
        print('La variable no es mayor a cero')
else:
    print('La variable no es un entero')
```

```python
#Resuelto con bucle For
numero = 5
factorial = 1
for i in range(2, (numero + 1)):
    factorial = i * factorial
print(factorial)
```

9) Imprimir los números primos existentes entre 0 y 30

```python
tope_rango = 30
n = 0
primo = True
while(n<tope_rango):
    for div in range(2, n):
        if n % div == 0:
            primo = False
    if primo:
        print(n)
    else:
        primo  = True

    n+=1
```

12) Aplicando continue, armar un ciclo while que solo imprima los valores divisibles por 12, dentro del rango de números de 100 a 300

```python
n = 99
while(n <= 300):
    n += 1
    if (n % 12 != 0):
        continue
    print(n, 'es divisible por 12')
```

```python
#Explicacion
#Inicializamos la variable n en 99
n = 99

# Comenzamos un bucle while que se ejecutara mientras n sea menor o igual a 300
while n <= 300:
    #incrementamos n en 1 para pasar al siguiente numero
    n += 1
    #comprobamos si n no es divisible por 12 utilizando el operador de modulo (%)
    if n % 12 != 0:
        # si n no es divisible por 12, se ejecuta la instruccion 'continue' para volver al inicio del bucle while
        # y saltar a la siguiente iteracion sin ejecutar el codigo que sigue
        continue

    # si se llega a esta linea de codigo, significa que n es divisible por 12
    # por lo que se imprime el mensaje indicando que n es divisible por 12
    print(n, 'es divisible por 12')
```

13) Utilizar la función **input()** que permite hacer ingresos por teclado, para encontrar números primos y dar la opción al usario de buscar el siguiente

```python
n = 1
sigue = 1
primo = True
while( sigue == 1):
    for div in range(2, n):
        if (n % div == 0):
            primo = False
            break
    if primo:
        print(n)
        print('¿Desea encontrar el siguiente numero primo?')
        if (input() != '1'):
            print('Se finaliza el proceso')
            break
    else:
        primo = True
    n += 1
    
```

```python
# inicializamos la variable n en 1
n = 1
# inicializamos la variable sigue en 1, esto indica si el proceso de busqueda continua
sigue = 1

#inicializamos la variable primo como verdadera
primo = True

#Comenzamos un bucle while que se ejecutara mientras sigue sea igual a 1
while sigue == 1:
    #comenzamos un bucle while que se ejecutara mientras sigue sea igual a 1
    while sigue == 1:
        #comenzamos un bucle for que itera desde 2 hasta n-1
        for div in range(2, n):
            #comprobamos si n es divisible entre div
            if n % div == 0:
                #si es divisible, marcamos primo como falso y salimos del bucle for
                primo = False
                break
        #verificamos si la variable primo es verdadera
        if primo:
            # si es verdadera, significa que n es un numero primo, por lo que lo imprimimos
            print(n)
            print('¿Desea encontrar el siguiente numero primo?' )

            #leemos la entrada del usuario para determinar si se continua o finaliza el proceso
            if input() != '1':
                print('Se finaliza el proceso')
                break
        else:
            #si primo es falso, significa que n no es un numero primo, por lo que lo restablecemos a verdadero
            primo = True

        # incrementamos n en 1 para pasar al siguiente numero
        n += 1
        
```

14) Crear un ciclo while que encuentre dentro del rango de 100 a 300 el primer número divisible por 3 y además múltiplo de 6

```python
n = 100
while(n<=300):
    if (n % 6 == 0):
        print('El numero es: ', str(n))
        break
    n += 1
```

```python
# Explicacion
# Inicializamos la variable n en 100
n = 100

# Comenzamos un bucle while que se ejecutará mientras n sea menor o igual a 300
while n <= 300:
    # Comprobamos si n es divisible por 6 sin dejar residuo
    if n % 6 == 0:
        # Si es divisible, imprimimos el mensaje y el valor de n
        print('El número es:', n)
        # Salimos del bucle while utilizando la instrucción 'break'
        break
    
    # Incrementamos n en 1 para pasar al siguiente número
    n += 1
    
```
